```python
import numpy as np
import matplotlib.pyplot as plt
from src.model import SimpleLSTM
from random import randint
```

```python
X = np.load("Data/NPY/X_unified.npy")
y = np.load("Data/NPY/y.npy")
```

```python
model = SimpleLSTM.load("best_simple_lstm.pth")
model.summary()
```

```python
index = randint(0, X.shape[0] - 1)
```

```python
def plot_predictions(model, X, y, index):
    preds = model.predict(X[index : index + 1])
    plt.figure(figsize=(12, 6))
    plt.plot(X[index, :, 0], X[index, :, 1], color="blue", label="Input Path")
    plt.scatter(
        y[index, 0], y[index, 1], color="red", label="Target Points", marker="x"
    )
    for point in preds:
        plt.scatter(
            point[0], point[1], color="green", label="Predicted Points", marker="o"
        )
    plt.xlabel("Latitude (degrees)")
    plt.ylabel("Longitude (degrees)")
    plt.legend()
    plt.show()


plot_predictions(model, X, y, index)
```
